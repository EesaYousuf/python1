```python
# var and std with respect to stastics lib
# data=[12,14,15,16,17,18,19,20]
# simple variance
def var(data_copy1):
    n=len(data_copy1)
    mean=sum(data_copy1)/n
    deviation=[(x-mean)**2 for x in data_copy1]
    variance=sum(deviation)/(n-1)
    return variance
```

```python
print(var(data_copy1))
```

```python
np.var(data_copy1)
```

```python
import statistics
statistics.variance(data_copy1)
```

```python
# population varience
def var(data_copy1):
    n=len(data_copy1)
    mean=sum(data_copy1)/n
    deviation=[(x-mean)**2 for x in data_copy1]
    variance=sum(deviation)/(n)
    return variance
```

```python
np.var(data_copy1)
```

```python
import statistics
statistics.variance(data_copy1)
```

```python

statistics.pvariance(data_copy1)
```

```python
sstd=var(data_copy1)
np.sqrt(sstd)
```
