```python
import seaborn as sns
from matplotlib import pyplot as plt
fmri=sns.load_dataset('fmri')
fmri.head()

```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
# change the style of the graph
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
# adding Markers
sns.lineplot(x='timepoint',y='signal',hue='event',style='event',markers=True,data=fmri)
```

```python
# Barplot
import pandas as pd
sns.set(style='whitegrid')
pokemon=pd.read_csv('Pokemon.csv')
pokemon.head()
```

```python
sns.barplot(x='Generation',y='Speed',data=pokemon,)
plt.show()
```

```python
sns.barplot(x='Sp. Def',y='Sp. Atk',data=pokemon)
plt.show()
```

```python
sns.barplot(x='Defense',y='HP',data=pokemon)
plt.show()
```

```python
sns.barplot(x='Generation',y='Speed',hue='Defense',data=pokemon)
plt.show()
```

```python
# seaborn scatterplot
iris=pd.read_csv('iris.csv')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
plt.show()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='class',style='class')
plt.show()
```
